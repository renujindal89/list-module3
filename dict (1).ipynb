# Looping Through a Dictionary

```python
student = {
    "name": "Aarav",
    "age": 21,
    "course": "Computer Science"
}
student.items()
```

```python
for key in student:
    print(key, ":", student[key])
```

```python
for key, value in student.items():
    print(key, "=", value)
```

```python
for key in student.keys():
    print(key)
```

```python
for value in student.values():
    print(value)
```

# merge the two dict

```python
#If keys overlap, values from the second dictionary overwrite the first
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}

merged = dict1 | dict2
print(merged)
```

```python
dict1.update(dict2) #dict1 is changed itself
print(dict1)
```

```python
merged = {k: dict1.get(k, 0) + dict2.get(k, 0) for k in set(dict1) | set(dict2)}
print(merged)
```
