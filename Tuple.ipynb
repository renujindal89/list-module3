# adding a new tuple

```python

In Python, tuples are immutable, which means you can’t directly add or remove elements 
once a tuple is created.
However, you can create a new tuple that includes the additional element(s).
#If you need to actually change the contents, convert to a list first.
```

# 1st method (create another tuple)

```python
#However, you can create a new tuple that includes the additional element(s).
#Using + (concatenation)
t = (1, 2, 3) # check the id
print(id(t))
t = t + (4,)
print(t)  # Output: (1, 2, 3, 4)
print(id(t))
```

```python
# second method (conversion into list)
```

```python
t1 = (1, 2, 3)
print(id(t1))
temp = list(t1)
temp.append(4)
t1 = tuple(temp)
print(id(t1))
print(t1)         #result store in  new tuple
```

# zip function

```python
 zip() function in Python is used to  pair up elements from two or more iterables
(like lists, tuples, or strings).
Each tuple contains one element from tuple1 and one from tuple2.
```

```python
t1 = (1, 2, 3)
t2 = ('a', 'b', 'c')
zip(t1, t2)
```

```python
t1 = (1, 2, 3)
t2 = ('a', 'b', 'c')
#zip(t1, t2)
a=zip(t1,t2)
print(a) # zip object (iterator)
print(list(a)) #reuse ,view the content,indexed
print(list(a))  #Zip Objects Are “Exhausted”

```

```python
#To fix it, store the zipped data in a list before you print or unzip it:
t1 = (1, 2, 3)
t2 = ('a', 'b', 'c')

zipped = list(zip(t1, t2))  # store the pairs
print(zipped)

print(zipped)
```

```python
a = (1, 2, 3)
b = ('x', 'y', 'z')
c = (True, False, True)

print(list(zip(a, b, c)))
```

```python
# Unequal Length Tuples
# zip() stops at the shortest one.
a = (1, 2, 3, 4)
b = ('a', 'b')

print(list(zip(a, b)))
```

# Unzipping (Reverse of Zip)

```python

#use the * operator to unpack the zipped result back into separate tuples.
t1 = (1, 2,3)
t2 = ('a', 'b', 'c')

zipped = list(zip(t1, t2))
print(zipped)
```

```python
t1,t2=zip(*zipped) # t3 error
print(t1)
print(t2)
```

# packing and unpacking 

```python
In Python, packing and unpacking 
group multiple values into a single variable (packing) and 
then extract them back into separate variables (unpacking).

# Packing
person = ("Alice", 25, "Engineer")

# Unpacking
name, age, job = person
print(name)  # Alice
print(age)   # 25
print(job)   # Engineer

a, *b = (1, 2, 3, 4)
print(a)  # 1
print(b)  # [2, 3, 4]

#convert list to tuple
t=tuple([1,2,'renu'])
t        

#convert string to tuple
t=tuple("renu")
t
```

# Converting 2D Tuple to List (for modification)

```python
matrix = (
    (1, 2),
    (3, 4)
)
matrix
```

```python
temp=[]
for row in matrix:
    temp.append(list(row))
print(temp)    
temp[0][1]=100
print(temp)
```

# convert back

```python
matrix=() #start with empty tuple
for row in temp:
    matrix = matrix + (tuple(row),)

print(matrix)
```
